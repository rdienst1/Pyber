```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "C:\\Users\\Rob\\Downloads\\city_data.csv"
ride_data_to_load = "C:\\Users\\Rob\\Downloads\\ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_complete_df = pd.merge(city_data_df, ride_data_df, how="left", on=["city"])

# Display the data table for preview
city_ride_complete_df
```

```python
#Calculate variables

#Average fare in each city

average_fare = city_ride_complete_df.groupby(['city'])['fare'].mean()
average_fare_df = pd.DataFrame(average_fare)
average_fare_df
```

```python
#Total rides per city

total_rides = city_ride_complete_df.groupby(['city'])['ride_id'].count()
total_rides_df = pd.DataFrame(total_rides)
total_rides_df
```

```python
#Total drivers per city

total_drivers = city_ride_complete_df.groupby(['city'])['driver_count'].unique()
total_drivers_df = pd.DataFrame(total_drivers)
total_drivers_df
```

```python
#City type

city_type = city_ride_complete_df.groupby(['city'])['type'].unique()
city_type_df = pd.DataFrame(city_type)
city_type_df
```

```python
#Combine all variables 

variables = [average_fare_df, total_rides_df, total_drivers_df, city_type_df]

#Concatenate into a single dataframe

variables_df = pd.concat(variables, sort=False, axis=1)

#Rename columns to be more representative of the dataset

variables_df.rename(columns={'fare':'average_fare','ride_id':'total_rides','driver_count':'total_drivers','type':'city_type'}, inplace=True)

variables_df
```

```python
#Build scatter plots for each city types; use dataframe plot method

colors = ["gold", "lightskyblue", "lightcoral"]
scatter_plot = variables_df.plot.scatter(x= 'total_rides', y= 'average_fare', c=colors, grid=True, figsize=(10,10), title= "Total Rides per City")                                        
scatter_plot.set(xlabel='Total Rides', ylabel='Average Fare in Dollars')

#plt.legend(['gold', 'lightskyblue', 'lightcoral'],['Urban', 'Suburban', 'Rural'])'

plt.savefig("C:\\Users\\Rob\\Documents\\PyberBubbleChart.png")
plt.show()
```

```python
#Calculate total fares by city type

fares_by_type = city_ride_complete_df.groupby(['type'])['fare'].sum()
fares_by_type_df = pd.DataFrame(fares_by_type)

#Calculate total of all fares

total_fares = fares_by_type_df['fare'].sum()

#Calculate percentages into dataframe

fares_by_type_df['percent'] = fares_by_type_df.loc[:,'fare':].sum(axis=1)/total_fares * 100
fares_by_type_df

```

```python
#Create pie chart

colors = ["gold", "lightskyblue", "lightcoral"]
fares_by_type_chart = fares_by_type_df.plot(kind="pie", y='fare', colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, title="Percentage of Fares by City Type")
fares_by_type_chart.axis("equal")
plt.savefig("C:\\Users\\Rob\\Documents\\FaresByType.png")
plt.show()
```
